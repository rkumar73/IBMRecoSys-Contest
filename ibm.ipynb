```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
traindf = pd.read_csv('train_5UKooLv.csv')
testdf = pd.read_csv('test_J1hm2KQ.csv')
testdf.shape,traindf.shape
```

```python
combdf = pd.concat([traindf,testdf],sort=False)
combdf.shape
```

```python
cols = ['CustomerID','StockCode']
combdf1 = combdf[cols].copy();combdf1.head()
```

```python
train_custid = traindf['CustomerID'].unique();len(train_custid)
```

```python
test_custid = testdf['CustomerID'].unique();len(test_custid)
```

```python
combtable = pd.pivot_table(combdf1,index=['CustomerID'],columns=['StockCode'],aggfunc=len, fill_value=0);combtable.shape
```

```python
combtable.head()
```

```python
test = combtable.loc[test_custid];test.shape
```

```python
train = combtable.loc[train_custid];train.shape
```

```python
kmeans = KMeans(n_jobs=-1,n_clusters=2).fit(train);(kmeans.labels_).shape
```

```python
clusters = kmeans.labels_
```

```python
train_cluster = train.copy()
train_cluster['cluster'] = clusters
train_cluster.shape
```

```python
clf = RandomForestClassifier(n_jobs=-1,min_samples_leaf=3,n_estimators=100,max_features=.5).fit(train,clusters)
pred = clf.predict(train)
(pred == clusters).mean()
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,clusters, test_size=0.1, random_state=42)
```

```python
cluster_table = pd.pivot_table(train_cluster,index=['cluster'],aggfunc=np.sum);cluster_table
```

```python
# This function gets the most popular clusters in the cell by grouping by the cluster column
def get_popular_products_in_cluster(cluster):
    plist = cluster_table.iloc[cluster]
    #print plist
    return plist.sort_values(ascending=False)
    #return plist
    
#topP =get_popular_products_in_cluster(19)
```

```python
topPlist = []
for i in range(len(cluster_table)):
    topPlist.append(cluster_table.iloc[i].sort_values(ascending=False))
```

```python
def find_product_reco(topP,already_pur_list):
    rec_list = []
    j=1
    for i in range(len(topP)):
    #for i in range(250):
        if  topP.index[i] in already_pur_list:
            #rec_list.insert(250-j,topP.index[i])
            #j = j+1
            #rec_list.append(topP.index[i])
            pass
            #print topP.index[i]
        else:
            rec_list.append(topP.index[i])
            
    return rec_list
```

```python
p = {}
for i in range(len(test)):
#for i in range(2):
    customer_cluster = test_cluster.iloc[i].cluster
    #print 'cust', customer_cluster
    #topProducts = get_popular_products_in_cluster(customer_cluster)
    topProducts = topPlist[customer_cluster]
    customer_row = test.iloc[i]
    already_bought_list = customer_row[customer_row>0].index.values
    recomendations_list = find_product_reco(topProducts,already_bought_list)
    p[test.index[i]] = dict(Items=recomendations_list[:250])

    #print test.index[i],customer_cluster,recomendations_list[:2]  #, already_bought_list
```

```python
s = pd.DataFrame.from_dict(p,orient='index',);s.index.name = 'CustomerID'
```
